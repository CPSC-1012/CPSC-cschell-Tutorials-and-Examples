1. Create an array of integers of size 100 that holds random numbers between 1 and 100 inclusive (in a method). The user will input a number to search for in the array. Create a method that, when passed in the array, will return the count of all numbers less than the search number. If there are no numbers less than the search number, return 0. Display the results in a method. Create a second method that, when passed in the array, will return the sum of all numbers greater than the search number. If there are no numbers greater than the search number, return 0. Display the results in a method.

```
using System;

namespace ConsoleApp36
{
    class Program
    {
        static void Main()
        {
            const int MAX_LENGTH = 100;
            int[] randomOneHundred = new int[MAX_LENGTH];
            Random random = new();
            for (int i = 0; i < randomOneHundred.Length; i++)
            {
                randomOneHundred[i] = random.Next(1, 101);
            }
            string prompt = "What number would you like to search for?: ";
            int searchTerm = GetPositiveInteger(prompt);

            DisplayCount(randomOneHundred, searchTerm);
            DisplaySum(randomOneHundred, searchTerm);
        }

        static int NumberLessThanSearchTerm(int[] numbers, int searchTerm)
        {
            int result = 0;
            foreach (int number in numbers)
            {
                if (number < searchTerm)
                {
                    result++;
                }
            }
            return result;
        }

        static void DisplayCount(int[] numbers, int searchTerm)
        {
            int result = NumberLessThanSearchTerm(numbers, searchTerm);
            if (result == 0)
            {
                Console.WriteLine("No numbers less than the search term were found in the array");
            }
            else
            {
                Console.WriteLine($"The count of numbers less than the search term is {result}");
            }
        }

        static int SumGreaterThanSearchTerm(int[] numbers, int searchTerm)
        {
            int sum = 0;
            foreach (int number in numbers)
            {
                if (number > searchTerm)
                {
                    sum += number;
                }
            }
            return sum;
        }

        static void DisplaySum(int[] numbers, int searchTerm)
        {
            int sum = SumGreaterThanSearchTerm(numbers, searchTerm);
            if (sum == 0)
            {
                Console.WriteLine("No numbers greater than the search term were found in the array");
            }
            else
            {
                Console.WriteLine($"The sum of numbers greater than the search term is {sum}");
            }
        }

        static int GetPositiveInteger(string prompt)
        {
            int result;
            Console.Write(prompt);
            while (!int.TryParse(Console.ReadLine(), out result) || result < 1)  
            {
                Console.WriteLine($"INVALID INPUT: {prompt}");
            }
            return result;
        }
    }
}
```
