# Question 1

```
using System;

namespace MethodExercisesOne
{
    class Program
    {
        static void DisplayMenu()
        {
            Console.WriteLine("Please select from the following options: ");
            Console.WriteLine("a - Enter your age");
            Console.WriteLine("b - Display admission cost");
            Console.WriteLine("x - Exit program");
            Console.Write("Your selection: ");
        } // end of method

        static int GetSafeInteger()
        {
            bool validInput = int.TryParse(Console.ReadLine(), out int userInput);
            while (!validInput)
            {
                Console.Write("Invalid input. Please enter a valid integer: ");
                validInput = int.TryParse(Console.ReadLine(), out userInput);
            } // end of while
            
            Console.WriteLine();
            
            return userInput;
        } // end of method

        static double GetTicketPrice(int age)
        {
            double ticketPrice;
            switch (age)
            {
                case <= 6:
                    ticketPrice = 0;
                    break;
                case < 18:
                    ticketPrice = 9.80;
                    break;
                case < 55:
                    ticketPrice = 11.35;
                    break;
                default:
                    ticketPrice = 10.00;
                    break;
            } // end of switch

            return ticketPrice; 
        } // end of method
        
        static void Main()
        {
            // declaring variables
            int age = 0;
            char userSelection;
            double admissionCost;

            do
            {
                DisplayMenu();
                userSelection = Console.ReadKey().KeyChar;

                Console.WriteLine();

                switch (userSelection)
                {
                    case 'a':
                        {
                            Console.Write("Please enter your age: ");
                            age = GetSafeInteger();
                            while (age < 1)
                            {
                                Console.Write("Please enter a valid age: ");
                                age = GetSafeInteger();
                            } // end of while
                            break;
                        } // end of case

                    case 'b':
                        {
                            admissionCost = GetTicketPrice(age);
                            
                            if (admissionCost == 0)
                            {
                                Console.WriteLine($"Your ticket is FREE!");
                            } // end of if
                            else
                            {
                                Console.WriteLine($"Your ticket price is {admissionCost:C}");
                            } // end of else
                            
                            Console.WriteLine("Thank you and enjoy the show!\n");
                            Console.Write("Press any key to return to the main menu: ");
                            Console.ReadKey();
                            break;
                        } // end of case

                    case 'x':
                        {
                            break;
                        } // end of case

                    default:
                        {
                            Console.WriteLine("Invalid menu choice. Please try again.");
                            break;
                        } // end of case
                } // end of switch

                Console.Clear();

            } while (userSelection != 'x');

            Console.WriteLine("\nThank you and don't forget to visit the concession!\n");

        } // end of method
    } // end of class
} // end of namespace
```

```
using System;

namespace MethodExerciesesTwo
{
    class Program
    {
        static int GetSafeInteger()
        {
            bool validInput = int.TryParse(Console.ReadLine(), out int userInput);
            while (!validInput)
            {
                Console.Write("Invalid input. Please enter a valid integer: ");
                validInput = int.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            Console.WriteLine();

            return userInput;

        } // end of method

        static char LetterGradeFromMark(int mark)
        {
            char letterGrade;
            switch (mark)
            {
                case < 50:
                    letterGrade = 'F';
                    break;
                case < 70:
                    letterGrade = 'D';
                    break;
                case < 80:
                    letterGrade = 'C';
                    break;
                case < 90:
                    letterGrade = 'B';
                    break;
                default:
                    letterGrade = 'A';
                    break;
            } // end of switch
            
            return letterGrade;

        } // end of method
        static void Main()
        {
            int grade;
            string name;
            char letterGrade;

            Console.Write("Please enter your name: ");
            name = Console.ReadLine();

            Console.WriteLine();

            Console.Write("Please enter your grade: ");
            grade = GetSafeInteger();

            while (grade < 0 || grade > 100)
            {
                Console.Write("Please enter a valid grade: ");
                grade = GetSafeInteger();
            } // end of while

            letterGrade = LetterGradeFromMark(grade);

            Console.WriteLine($"{name}, your grade is a {letterGrade}");
        } // end of method
    } // end of class
} // end of namespace
```

```
using System;

namespace MethodExercisesThree
{
    class Program
    {
        static double GetSafeDouble()
        {
            bool validInput = double.TryParse(Console.ReadLine(), out double userInput);
            while (!validInput)
            {
                Console.Write("Invalid input. Please enter a valid number: ");
                validInput = double.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            Console.WriteLine();

            return userInput;
        } // end of method

        static double TaxesOwed(double income)
        {
            double incomeTax;
            if (income < 50000)
            {
                incomeTax = income * 0.05;
            } // end of if
            else if (income < 100000)
            {
                incomeTax = 2500 + (income - 50000) * 0.07;
            } // end of else if
            else
            {
                incomeTax = 6000 + (income - 100000) * 0.09;
            } // end of else
            
            return incomeTax;
        } // end of method
        static void Main()
        {
            double incomeAmount;
            double incomeTax;

            Console.Write("Please provide your income: $");
            incomeAmount = GetSafeDouble();

            while (incomeAmount < 0)
            {
                Console.Write("Please provide a valid income: $");
                incomeAmount = GetSafeDouble();
            } // end of while

            incomeTax = TaxesOwed(incomeAmount);

            Console.WriteLine($"The income tax due based on an income of {incomeAmount:C} is {incomeTax:C}");
        } // end of method
    } // end of class
} // end of namespace
```
