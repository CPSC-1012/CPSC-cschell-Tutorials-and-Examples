# `Program.cs`

```
using System;

namespace CalculateGrades
{
    internal class Program
    {
        static void Main()
        {
            Lab labs = new();
            Quiz quizzes = new();
            Assignment assignment = new();

            string firstName = WelcomeAndNameCapture();

            bool isCorrect = labs.CaptureLabs(firstName);
            if (!isCorrect)
            {
                labs.FixLabs(firstName);
            }

            isCorrect = quizzes.CaptureQuizzes(firstName);
            if (!isCorrect)
            {
                quizzes.FixQuizzes(firstName);
            }

            isCorrect = assignment.CaptureAssignments(firstName);
            if (!isCorrect)
            {
                assignment.FixAssignments(firstName);
            }

        }

        static string WelcomeAndNameCapture()
        {
            Console.WriteLine(" Welcome to the CPSC1012 grade assistant!");
            Console.WriteLine("+----------------------------------------+");
            string message = "Can you please provide your first name?: ";
            return PromptForStringNotEmpty(message);
        }

        static string PromptForStringNotEmpty(string message)
        {
            string stringValue;
            bool validInput = false;

            do
            {
                Console.Write(message);
                stringValue = Console.ReadLine();
                if (!string.IsNullOrWhiteSpace(stringValue))
                {
                    validInput = true;
                }
                else
                {
                    Console.WriteLine("Invalid input value. Please enter a non-empty string.");
                }
            } while (!validInput);

            return stringValue;
        }
    }
}
```

# `Lab.cs`

```
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace CalculateGrades
{
    internal class Lab
    {
        private double _labOne;
        private double _labTwo;
        private double _labThree;   
        private double _labFour;
        private double _labFive;
        private double _labSix;

        public double LabOne
        {
            get 
            { 
                return _labOne; 
            }
            set 
            {
                if (value < 0 || value > 5)
                {
                    throw new Exception("Invalid mark: Lab 1 is out of 5");
                }
                else
                {
                    _labOne = value;

                }
            }
        }

        public double LabTwo
        {
            get 
            { 
                return _labTwo; 
            }
            set
            {
                if (value < 0 || value > 5)
                {
                    throw new Exception("Invalid mark: Lab 2 is out of 5");
                }
                else
                {
                    _labTwo = value;
                }
            }
        }

        public double LabThree
        {
            get
            {
                return _labThree;
            }
            set
            {
                if (value < 0 || value > 5)
                {
                    throw new Exception("Invalid mark: Lab 3 is out of 5");
                }
                else
                {
                    _labThree = value;
                }
            }
        }

        public double LabFour
        {
            get
            {
                return _labFour;
            }
            set
            {
                if (value < 0 || value > 5)
                {
                    throw new Exception("Invalid mark: Lab 4 is out of 5");
                }
                else
                {
                    _labFour = value;
                }
            }
        }

        public double LabFive
        {
            get
            {
                return _labFive;
            }
            set
            {
                if (value < 0 || value > 5)
                {
                    throw new Exception("Invalid mark: Lab 5 is out of 5");
                }
                else
                {
                    _labFive = value;
                }
            }
        }

        public double LabSix
        {
            get
            {
                return _labSix;
            }
            set
            {
                if (value < 0 || value > 5)
                {
                    throw new Exception("Invalid mark: Lab6 is out of 5");
                }
                else
                {
                    _labSix = value;
                }
            }
        }

        public Lab()
        {

        }

        public bool CaptureLabs(string firstName)
        {
            Console.Clear();
            Console.Write($"{firstName}, please enter your Lab 1 mark out of five (0 if not yet complete): ");
            LabOne = GetSafeLab();
            Console.Write($"{firstName}, please enter your Lab 2 mark out of five (0 if not yet complete): ");
            LabTwo = GetSafeLab();
            Console.Write($"{firstName}, please enter your Lab 3 mark out of five (0 if not yet complete): ");
            LabThree = GetSafeLab();
            Console.Write($"{firstName}, please enter your Lab 4 mark out of five (0 if not yet complete): ");
            LabFour = GetSafeLab();
            Console.Write($"{firstName}, please enter your Lab 5 mark out of five (0 if not yet complete): ");
            LabFive = GetSafeLab();
            Console.Write($"{firstName}, please enter your Lab 6 mark out of five (0 if not yet complete): ");
            LabSix = GetSafeLab();

            Console.WriteLine(ToString());

            Console.Write($"Are the entered lab marks correct, {firstName}? (y or n): ");
            char yesNo = char.ToLower(Console.ReadKey().KeyChar);
            Console.WriteLine();
            if (yesNo == 'y')
            {
                return true;
            }
            else if (yesNo == 'n')
            {
                return false;
            }
            else
            {
                Console.WriteLine($"I will take {yesNo} as a no, {firstName} ...");
                return false;
            }
        }

        static int GetSafeLabInput()
        {
            bool validInput = int.TryParse(Console.ReadLine(), out int userInput);
            while (!validInput || userInput < 0 || userInput > 6)
            {
                Console.Write("Invalid input. Please enter a valid integer: ");
                validInput = int.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            Console.WriteLine();

            return userInput;
        }

        static double GetSafeLab()
        {
            bool validInput = double.TryParse(Console.ReadLine(), out double userInput);
            while (!validInput || userInput < 0 || userInput > 5)
            {
                Console.Write("Invalid. Labs are out of 5 marks: ");
                validInput = double.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            return userInput;
        }

        public void FixLabs(string firstName)
        {
            int userInput;
            do
            {
                Console.Clear();
                Console.WriteLine($"1 - Lab 1: {LabOne}/5");
                Console.WriteLine($"2 - Lab 2: {LabTwo}/5");
                Console.WriteLine($"3 - Lab 3: {LabThree}/5");
                Console.WriteLine($"4 - Lab 4: {LabFour}/5");
                Console.WriteLine($"5 - Lab 5: {LabFive}/5");
                Console.WriteLine($"6 - Lab 6: {LabSix}/5");
                Console.WriteLine($"0 - Exit");
                Console.Write($"Which Lab do you need to adjust, {firstName}?: ");
                userInput = GetSafeLabInput();

                switch (userInput)
                {
                    case 1:
                        {
                            Console.Write("What is the accurate Lab 1 mark?: ");
                            LabOne = GetSafeLab();
                            break;
                        }
                    case 2:
                        {
                            Console.Write("What is the accurate Lab 2 mark?: ");
                            LabTwo = GetSafeLab();
                            break;
                        }
                    case 3:
                        {
                            Console.Write("What is the accurate Lab 3 mark?: ");
                            LabThree = GetSafeLab();
                            break;
                        }
                    case 4:
                        {
                            Console.Write("What is the accurate Lab 4 mark?: ");
                            LabFour = GetSafeLab();
                            break;
                        }
                    case 5:
                        {
                            Console.Write("What is the accurate Lab 5 mark?: ");
                            LabFive = GetSafeLab();
                            break;
                        }
                    case 6:
                        {
                            Console.Write("What is the accurate Lab 6 mark?: ");
                            LabSix = GetSafeLab();
                            break;
                        }
                    case 0:
                        {
                            Console.WriteLine($"Thanks {firstName}, looks good now!");
                            break;
                        }
                    default:
                        {
                            Console.WriteLine($"Hello {firstName}, please make another selection: ");
                            break;
                        }
                }
            } while (userInput != 0);

            Console.WriteLine(ToString());
            Console.Write($"Thanks for the edits {firstName}! Are you ready to proceed? (press any key): ");
            Console.ReadKey();
        }

        public override string ToString()
        {
            double labGrade = LabOne + LabTwo + LabThree + LabFour + LabFive + LabSix;
            string labOutput;
            if (labGrade < 15)
            {
                labOutput = $"You require {Math.Round(15 - labGrade, 1)} additional percentage points to pass.";
            }
            else
            {
                labOutput = $"You already are passing your labs with {Math.Round(labGrade, 1)}/30 percentage points.";
                
            }
            return $"LAB    MARK/5\tLAB %\t\tPERCENTAGE TOWARD FINAL\n" +
                $"-----------------------------------------------------------------------\n" +
                $"Lab 1: {LabOne}/5\t{Math.Round(LabOne / 5, 1):P1}\t\t{LabOne}% toward final grade.\n" +
                $"Lab 2: {LabTwo}/5\t{LabTwo / 5:P1}\t\t{LabTwo}% toward final grade.\n" +
                $"Lab 3: {LabThree}/5\t{LabThree / 5:P1}\t\t{LabThree}% toward final grade.\n" +
                $"Lab 4: {LabFour}/5\t{LabFour / 5:P1}\t\t{LabFour}% toward final grade.\n" +
                $"Lab 5: {LabFive}/5\t{LabFive / 5:P1}\t\t{LabFive}% toward final grade.\n" +
                $"Lab 6: {LabSix}/5\t{LabSix / 5:P1}\t\t{LabSix}% toward final grade.\n" +
                $"{labOutput}";
        }
    }
}
```

# `Quiz.cs`

```
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace CalculateGrades
{
    internal class Quiz
    {
        private double _quizOne;
        private double _quizTwo;

        public double QuizOne
        {
            get 
            { 
                return _quizOne; 
            }
            set 
            {
                if (value < 0 || value > 45)
                {
                    throw new Exception("Invalid Quiz One Mark: Quiz is out of 45 marks.");
                }
                else
                {
                    _quizOne = value;
                }
            }
        }

        public double QuizTwo
        {
            get
            {
                return _quizTwo;
            }
            set
            {
                if (value < 0 || value > 43)
                {
                    throw new Exception("Invalid Quiz Two Mark: Quiz is out of 43 marks.");
                }
                else
                {
                    _quizTwo = value;
                }
            }
        }

        public Quiz()
        {
        }

        //public double CalculateQuizGrade()
        //{
        //    double quizTotal = (QuizOne / 45 * 10) + (QuizTwo / 43 * 10);
        //    return quizTotal;
        //}

        static double GetSafeQuiz(int quizNumber)
        {
            if (quizNumber == 1)
            {
                bool validInput = double.TryParse(Console.ReadLine(), out double userInput);
                while (!validInput || userInput < 0 || userInput > 45)
                {
                    Console.Write("Invalid. Quiz 1 is out of 45 marks: ");
                    validInput = double.TryParse(Console.ReadLine(), out userInput);
                } // end of while

                return userInput;
            }
            else
            {
                bool validInput = double.TryParse(Console.ReadLine(), out double userInput);
                while (!validInput || userInput < 0 || userInput > 43)
                {
                    Console.Write("Invalid. Quiz 2 is out of 43 marks: ");
                    validInput = double.TryParse(Console.ReadLine(), out userInput);
                } // end of while

                return userInput;
            }
        }

        static int GetSafeQuizInput()
        {
            bool validInput = int.TryParse(Console.ReadLine(), out int userInput);
            while (!validInput || userInput < 0 || userInput > 2)
            {
                Console.Write("Invalid input. Please enter a valid integer: ");
                validInput = int.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            Console.WriteLine();

            return userInput;
        }

        public bool CaptureQuizzes(string firstName)
        {
            Console.Clear();
            Console.Write($"{firstName}, please enter your Quiz 1 mark out of forty five (0 if not yet complete): ");
            QuizOne = GetSafeQuiz(1);
            Console.Write($"{firstName}, please enter your Quiz 2 mark out of forty three (0 if not yet complete): ");
            QuizTwo = GetSafeQuiz(2);

            Console.WriteLine(ToString());

            Console.Write($"Are the entered quiz marks correct, {firstName}? (y or n): ");
            char yesNo = char.ToLower(Console.ReadKey().KeyChar);
            Console.WriteLine();
            if (yesNo == 'y')
            {
                return true;
            }
            else if (yesNo == 'n')
            {
                return false;
            }
            else
            {
                Console.WriteLine($"I will take {yesNo} as a no, {firstName} ...");
                return false;
            }

        }

        public void FixQuizzes(string firstName)
        {
            int userInput;
            do
            {
                Console.Clear();
                Console.WriteLine($"1 - Quiz 1: {QuizOne}/45");
                Console.WriteLine($"2 - Quiz 2: {QuizTwo}/43");
                Console.WriteLine($"0 - Exit");
                Console.Write($"Which Quiz do you need to adjust, {firstName}?: ");
                userInput = GetSafeQuizInput();

                switch (userInput)
                {
                    case 1:
                        {
                            Console.Write("What is the accurate Quiz 1 mark?: ");
                            QuizOne = GetSafeQuiz(1);
                            break;
                        }
                    case 2:
                        {
                            Console.Write("What is the accurate Quiz 2 mark?: ");
                            QuizTwo = GetSafeQuiz(2);
                            break;
                        }
                    case 0:
                        {
                            Console.WriteLine($"Thanks {firstName}, looks good now!");
                            break;
                        }
                    default:
                        {
                            Console.WriteLine($"Hello {firstName}, please make another selection: ");
                            break;
                        }
                }
            } while (userInput != 0);

            Console.WriteLine(ToString());
            Console.Write($"Thanks for the edits {firstName}! Are you ready to proceed? (press any key): ");
            Console.ReadKey();
        }

        public override string ToString()
        {
            double quizGrade = (QuizOne / 45 * 10) + (QuizTwo / 43 * 10);
            string quizOutput;
            if (quizGrade < 10)
            {
                quizOutput = $"You require {Math.Round(10 - quizGrade, 1)} additional percentage points to pass.";
            }
            else
            {
                quizOutput = $"You already are passing your quizzes with {Math.Round(quizGrade, 1)}/20 percentage points.";

            }
            return $"QUIZ    MARK\tQUIZ %\t\tPERCENTAGE TOWARD FINAL\n" +
                $"-----------------------------------------------------------------------\n" +
                $"Quiz 1: {QuizOne}/45\t{QuizOne / 45:P1}\t\t{Math.Round(QuizOne / 45 * 10, 1)}% toward final grade.\n" +
                $"Quiz 2: {QuizTwo}/43\t{QuizTwo / 43:P1}\t\t{Math.Round(QuizTwo / 43 * 10, 1)}% toward final grade.\n" +
                $"{quizOutput}";
        }
    }
}
```

```
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;

namespace CalculateGrades
{
    internal class Assignment
    {
        private double _assignmentOne;
        private double _assignmentTwo;
        private double _assignmentThree;
        private double _assignmentFour;

        public double AssignmentOne
        { 
            get 
            { 
                return _assignmentOne; 
            } 
            set 
            {
                if (value < 0 || value > 100)
                {
                    throw new Exception("Invalid Assignment Mark: Assignment is out of 100 marks.");
                }
                else
                {
                    _assignmentOne = value;
                }
            } 
        }

        public double AssignmentTwo
        {
            get
            {
                return _assignmentTwo;
            }
            set
            {
                if (value < 0 || value > 100)
                {
                    throw new Exception("Invalid Assignment Mark: Assignment is out of 100 marks.");
                }
                else
                {
                    _assignmentTwo = value;
                }
            }
        }

        public double AssignmentThree
        {
            get
            {
                return _assignmentThree;
            }
            set
            {
                if (value < 0 || value > 100)
                {
                    throw new Exception("Invalid Assignment Mark: Assignment is out of 100 marks.");
                }
                else
                {
                    _assignmentThree = value;
                }
            }
        }

        public double AssignmentFour
        {
            get
            {
                return _assignmentFour;
            }
            set
            {
                if (value < 0 || value > 100)
                {
                    throw new Exception("Invalid Assignment Mark: Assignment is out of 100 marks.");
                }
                else
                {
                    _assignmentFour = value;
                }
            }
        }

        public Assignment()
        {
        }

        //public double CalculateAssignmentGrade()
        //{
        //    double assignmentGrade;
        //    assignmentGrade = (AssignmentOne / 100 * 10) + (AssignmentTwo / 100 * 10) + (AssignmentThree / 100 * 10) + (AssignmentFour / 100 * 20);
        //    return assignmentGrade;
        //}

        static double GetSafeAssignment()
        {
            bool validInput = double.TryParse(Console.ReadLine(), out double userInput);
            while (!validInput || userInput < 0 || userInput > 100)
            {
                Console.Write("Invalid. Assingment is out of 100 marks: ");
                validInput = double.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            return userInput;
        }

        static int GetSafeAssignmentInput()
        {
            bool validInput = int.TryParse(Console.ReadLine(), out int userInput);
            while (!validInput || userInput < 0 || userInput > 4)
            {
                Console.Write("Invalid input. Please enter a valid integer: ");
                validInput = int.TryParse(Console.ReadLine(), out userInput);
            } // end of while

            Console.WriteLine();

            return userInput;
        }

        public bool CaptureAssignments(string firstName)
        {
            Console.Clear();
            Console.Write($"{firstName}, please enter your Assignment 1 mark out of one hundred (0 if not yet complete): ");
            AssignmentOne = GetSafeAssignment();
            Console.Write($"{firstName}, please enter your Assignment 2 mark out of one hundred (0 if not yet complete): ");
            AssignmentTwo = GetSafeAssignment();
            Console.Write($"{firstName}, please enter your Assignment 3 mark out of one hundred (0 if not yet complete): ");
            AssignmentThree = GetSafeAssignment();
            Console.Write($"{firstName}, please enter your Assignment 4 mark out of one hundred (0 if not yet complete): ");
            AssignmentFour = GetSafeAssignment();

            Console.WriteLine(ToString());

            Console.Write($"Are the entered assignment marks correct, {firstName}? (y or n): ");
            char yesNo = char.ToLower(Console.ReadKey().KeyChar);
            Console.WriteLine();
            if (yesNo == 'y')
            {
                return true;
            }
            else if (yesNo == 'n')
            {
                return false;
            }
            else
            {
                Console.WriteLine($"I will take {yesNo} as a no, {firstName} ...");
                return false;
            }

        }

        public void FixAssignments(string firstName)
        {
            int userInput;
            do
            {
                Console.Clear();
                Console.WriteLine($"1 - Assignment 1: {AssignmentOne}/100");
                Console.WriteLine($"2 - Assignment 2: {AssignmentTwo}/100");
                Console.WriteLine($"3 - Assignment 3: {AssignmentThree}/100");
                Console.WriteLine($"4 - Assignment 4: {AssignmentFour}/100");
                Console.WriteLine($"0 - Exit");
                Console.Write($"Which Assignment do you need to adjust, {firstName}?: ");
                userInput = GetSafeAssignmentInput();

                switch (userInput)
                {
                    case 1:
                        {
                            Console.Write("What is the accurate Assignment 1 mark?: ");
                            AssignmentOne = GetSafeAssignment();
                            break;
                        }
                    case 2:
                        {
                            Console.Write("What is the accurate Assignment 2 mark?: ");
                            AssignmentTwo = GetSafeAssignment();
                            break;
                        }
                    case 3:
                        {
                            Console.Write("What is the accurate Assignment 3 mark?: ");
                            AssignmentThree = GetSafeAssignment();
                            break;
                        }
                    case 4:
                        {
                            Console.Write("What is the accurate Assignment 4 mark?: ");
                            AssignmentFour = GetSafeAssignment();
                            break;
                        }
                    case 0:
                        {
                            Console.WriteLine($"Thanks {firstName}, looks good now!");
                            break;
                        }
                    default:
                        {
                            Console.WriteLine($"Hello {firstName}, please make another selection: ");
                            break;
                        }
                }
            } while (userInput != 0);

            Console.WriteLine(ToString());
            Console.Write($"Thanks for the edits {firstName}! Are you ready to proceed? (press any key): ");
            Console.ReadKey();
        }

        public override string ToString()
        {
            double assignmentGrade = (AssignmentOne / 100 * 10) + (AssignmentTwo / 100 * 10) + (AssignmentThree / 100 * 10) + (AssignmentFour / 100 * 20);
            string assignmentOutput;
            if (assignmentGrade < 25)
            {
                assignmentOutput = $"You require {Math.Round(25 - assignmentGrade, 1)} additional percentage points to pass.";
            }
            else
            {
                assignmentOutput = $"You already are passing your assignments with {Math.Round(assignmentGrade, 1)}/50 percentage points.";

            }
            return $"ASSIGN    MARK/5\tASSIGN %\tPERCENTAGE TOWARD FINAL\n" +
                $"-----------------------------------------------------------------------\n" +
                $"Assign 1: {AssignmentOne}/100\t{AssignmentOne / 100:P1}\t\t{Math.Round(AssignmentOne / 100 * 10, 1)}% toward final grade.\n" +
                $"Assign 2: {AssignmentTwo}/100\t{AssignmentTwo / 100:P1}\t\t{Math.Round(AssignmentTwo / 100 * 10, 1)}% toward final grade.\n" +
                $"Assign 3: {AssignmentThree}/100\t{AssignmentThree / 100:P1}\t\t{Math.Round(AssignmentThree / 100 * 10, 1)}% toward final grade.\n" +
                $"Assign 4: {AssignmentFour}/100\t{AssignmentFour / 100:P1}\t\t{Math.Round(AssignmentFour / 100 * 20, 1)}% toward final grade.\n" +
                $"{assignmentOutput}";
        }
    }
}
```
