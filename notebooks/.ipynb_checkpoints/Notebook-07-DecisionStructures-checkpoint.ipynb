# Question 1
* Write a program that lets the user guess whether the flip of a coin results in heads or tails. The program randomly generates an integer 0 to 1, which represents head or tail. The program prompts the user to enter a guess, and reports whether the guess is correct or incorrect.

````
int coinFlip, userSelect;
Random randomFlip = new();
coinFlip = randomFlip.Next(2);

Console.WriteLine("Feeling lucky? Choose heads (1) or tails (0):");
/*
userSelect = int.Parse(Console.ReadLine());
*/
userSelect = 1;
if (userSelect == coinFlip)
{
    Console.WriteLine("Congratulations, you are correct!\n" +
        $"House: {coinFlip} Your choice: {userSelect}");
} // end of if
else
{
    Console.WriteLine("No worries! Statistically, the house \"always\" wins.\n" +
        $"House: {coinFlip} Your choice: {userSelect}");
} // end of else
````

# Question 2
* Write a program to determine if a user input year is a leap year. A year is a leap year if it is divisible by 4 but not by 100, or if it is divisible by 400. Here are some sample runs:

```
Enter a year:
2008
2008 is a leap year? true
```

----

```
Enter a year:
1900
1900 is a leap year? false
```

````
int year;

Console.Write("Enter a year: ");

// int year = int.Parse(Console.ReadLine());
year = 1985;

// Check if the year is a leap year
// A leap year is divisible by 4:
bool isLeapYear = (year % 4 == 0);
// A leap year is divisible by 4 but not by 100:
isLeapYear = isLeapYear && (year % 100 != 0);
// A leap year is divisible by 4 or 400 but not by 100:
isLeapYear = isLeapYear || (year % 400 == 0);

// combined:
// bool isLeapYear= (year % 4 == 0 && year % 100 != 0) || (year % 400 == 0);

// Display the result
Console.WriteLine($"{year} is a leap year? {isLeapYear}");
````

# Question 3
* Write a program that plays the scissor-rock-paper game. (A scissor cuts a paper, a rock can knock a scissor, and a paper can wrap a rock.) The program randomly generates a number 0, 1, or 2 representing scissor, rock, and paper. The program prompts the user to enter a number 0, 1, or 2 and display a message indicating whether the user or the computer wins, loses, or draws. Here are sample runs:

```
Scissor (0), rock (1), paper (2):
1
The computer is scissor. You are rock. You win.
```
----
```
Scissor (0), rock (1), paper (2):
1
The computer is paper. You are too. It is a draw.
```

````
// Variable
Random randomNumber = new();
int computerNumber = randomNumber.Next(3);
int userNumber;

// Input
Console.WriteLine("Let's play rock paper scissors. ");
Console.WriteLine("Rules: \n\t - scissor = 0; rock = 1; paper = 2;" +
                  "\n\t - scissor cuts a paper; rock can knock a scissor; paper can wrap a rock;");
Console.Write("\nEnter number 0, 1, or 2:\n\n");

// userNumber = int.Parse(Console.ReadLine());

userNumber = 1;

if (userNumber >= 0 && userNumber <= 2)
{
    if (userNumber == computerNumber)
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"Draw");
    }
    else if (userNumber == 0 && computerNumber == 1)
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"Computer Wins");
    }
    else if (userNumber == 0 && computerNumber == 2)
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"User Wins");
    }
    else if (userNumber == 1 && computerNumber == 2)
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"Computer Wins");
    }
    else if (userNumber == 1 && computerNumber == 0)
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"User Wins");
    }
    else if (userNumber == 2 && computerNumber == 1)
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"User Wins");
    }
    else
    {
        Console.WriteLine($"Computer plays {computerNumber} " +
            $"and user plays {userNumber} \n" +
            $"Computer Wins");
    }
}
else
{
    Console.WriteLine("Not a valid input. Please try again.");
} // end else
````

### Solution 2 - Question 3

````
//declare variables
int number;
string userChoice, computerChoice;

//Prompt users for input
Console.WriteLine("Please enter 0, 1, or 2:\n\tScissors(0), Rock(1), or Paper(2):\n\n");
// number = int.Parse(Console.ReadLine());
number = 2;

//generate computer for random numbers
Random rnd = new();
int computerRandom = rnd.Next(3);

//evaluation
if (number == computerRandom)
{ 
    Console.Write($"Draw:\nBoth chose ");
} // end if
else if (number == 0 && computerRandom == 2)
{
    Console.Write($"Win:\nThe computer chose Paper. You chose ");
} // end else if
else if (number == 2 && computerRandom == 0)
{
    Console.Write($"Loss:\nThe computer chose Scissors. You chose ");
} // end else if
else if (number > computerRandom)
{
    // set string data
    if (computerRandom == 0)
    {
        computerChoice = "Scissors";
    } // end if
    else if (computerRandom == 1)
    {
        computerChoice = "Rock";
    } // end else if
    else
    {
        computerChoice = "Paper";
    } // end else
    
    //Output user message
    Console.Write($"Win:\nThe computer chose {computerChoice}. You chose ");
}
else
{
    // set string data
    if (computerRandom == 0)
    {
        computerChoice = "Scissors";
    } // end if
    else if (computerRandom == 1)
    {
        computerChoice = "Rock";
    } // end else if
    else
    {
        computerChoice = "Paper";
    } // end else 
    
    //Output user message
    Console.Write($"Loss:\nThe computer chose {computerChoice}. You chose ");
}

// set string data
if (number == 0)
{
    userChoice = "Scissors";
} // end if
else if (number == 1)
{
    userChoice = "Rock";
} // end else if
else
{
    userChoice = "Paper";
} // end else

// finalize the output message
Console.Write(userChoice + ".");
````

# Question 4
* Write a program that reads in the taxable income of a person and computes the Canadian federal income tax rate and tax amount for a single person. The [federal tax rates for 2021](https://www.canada.ca/en/financial-consumer-agency/services/financial-toolkit/taxes/taxes-2/5.html) are:
    * 15% on the first $48,535 of taxable income
    * 20.5% on the next $48,534 of taxable income (on the portion of taxable income over $48,535 up to $97,069)
    * 26% on the next $53,404 of taxable income (on the portion of taxable income over $97,069 up to $150,473)
    * 29% on the next $63,895 of taxable income (on the portion of taxable income over $150,473 up to
$214,368)
    * 33% of taxable income over $214,368.

* The following table contains the formula for computing a single person’s tax (Federal tax rates for 2017).

| Bracket 	| Taxable Income       	| Marginal Tax Rate 	| Tax Paid                                 	|
|---------	|----------------------	|:-----------------:	|------------------------------------------	|
| 1       	| \$48,535 or less      	| 15%               	| 15%                                      	|
| 2       	| \$48,535 to \$97,069   	| 20.5%             	| \$7,280.25 + 20.5% on amount over \$48,535 	|
| 3       	| \$97,069 to \$150,473  	| 26%               	| \$17,229.72 + 26% on amount over \$97,069  	|
| 4       	| \$150,473 to \$214,368 	| 29%               	| \$31,114.76 + 29% on amount over \$150,473 	|
| 5       	| More than \$214,368   	| 33%               	| \$49,644.31 + 33% on amount over \$214,368 	|


* Here is a sample run:

```
If your taxable income was $50,000 in 2020
// Pay 15% on the amount up to $48,535, or $7,280.25
// Pay 20.5% on the amount between $48,535 to $97,069, or $300.33
Total federal tax payable: $7,580.58.
```


````
double income, totalTax;

Console.WriteLine("Please enter your income for 2020:");
// income = double.Parse(Console.ReadLine());
income = 65000;

switch (income)
{
    case <= 48535:
        totalTax = income * 0.15;
        break;
    case < 97069:
        totalTax = 7280.25 + (income - 48535) * 0.205;
        break;
    case < 150473:
        totalTax = 17229.72 + (income - 97069) * 0.26;
        break;
    case < 214368:
        totalTax = 31114.76 + (income - 150473) * 0.29;
        break;
    default:
        totalTax = 49644.31 + (income - 214368) * 0.33;
        break;
} // end switch

Console.WriteLine($"If your taxable income was {income:C} in 2020\n" +
    $"Your total federal tax payable is {totalTax:C}");
````

# Question 5
* Write a program that reads the taxable income of an individual and computes the province tax rate and tax amount for an individual in the province of Alberta. The Alberta provincial tax rates for 2017 are:

    * **10%** Up to \$131,220
    * **12%** \$131.220.01 to \$157,464
    * **13%** \$157,464.01 to \$209,952
    * **14%** \$209,952.01 to \$314,928
    * **15%** \$314,928.01 and up

Here is a sample run:

```
Enter your taxable income:
65000
Your Alberta provincial government tax rate is 10% and you owe the province $6,500.00 in taxes!
```

````
double income, totalTax;

Console.WriteLine("Please enter your income for 2020:");
// income = double.Parse(Console.ReadLine());
income = 65000;

switch (income)
{
    case <= 131220:
        totalTax = income * 0.10;
        Console.Write("Your Alberta provincial government tax rate is 10%\n");
        break;
    case < 157464:
        totalTax = income * 0.12;
        Console.Write("Your Alberta provincial government tax rate is 12%\n");
        break;
    case < 209952:
        totalTax = income * 0.13;
        Console.Write("Your Alberta provincial government tax rate is 13%\n");
        break;
    case < 314928:
        totalTax = income * 0.14;
        Console.Write("Your Alberta provincial government tax rate is 14%\n");
        break;
    default:
        totalTax = income * 0.15;
        Console.Write("Your Alberta provincial government tax rate is 15%\n");
        break;
} // end switch

Console.WriteLine($"If your taxable income was {income:C} in 2020\n" +
    $"Your total provincial tax payable is {totalTax:C}");
````

# Question 7
* Write a program that simulates picking a card from a deck of 52 cards. Your program should display the rank (Ace, 2, 3, 4, 5, 6, 7, 8, 9, 10, Jack, Queen, King) and the suit (Clubs, Diamonds, Hearts, Spades) of the card. Here is a sample of the program:

```
The card you picked is Jack of Hearts
```

````
// assign random ints and declare the variables
Random keygen = new Random();
int randomRank = keygen.Next(1, 14);
int randomSuit = keygen.Next(1, 5);
string cardSuit = "Suit";
string cardRank = "Rank";

// determine rank
switch (randomRank)
{
    case 1:
        cardRank = "Ace";
        break;
    case 2:
        cardRank = "2";
        break;
    case 3:
        cardRank = "3";
        break;
    case 4:
        cardRank = "4";
        break;
    case 5:
        cardRank = "5";
        break;
    case 6:
        cardRank = "6";
        break;
    case 7:
        cardRank = "7";
        break;
    case 8:
        cardRank = "8";
        break;
    case 9:
        cardRank = "9";
        break;
    case 10:
        cardRank = "10";
        break;
    case 11:
        cardRank = "Jack";
        break;
    case 12:
        cardRank = "Queen";
        break;
    case 13:
        cardRank = "King";
        break;
}    // end switch

// determine suit
switch (randomSuit)
{
    case 1:
        cardSuit = "Hearts";
        break;
    case 2:
        cardSuit = "Spades";
        break;
    case 3:
        cardSuit = "Diamonds";
        break;
    case 4:
        cardSuit = "Clubs";
        break;
} // end switch

// output values to console
Console.WriteLine($"The card you have drawn is the {cardRank} of {cardSuit}!");
````
