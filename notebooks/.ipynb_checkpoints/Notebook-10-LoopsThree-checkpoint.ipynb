# Question 1
A bacteriologist determines that the approximate final population of bacteria present in a culture after time (in days) is given by the following formula: 

FinalPopulation = InitialPopulation * e^(GrowthRate * Time) 

where InitialPopulation is the number present at the beginning of the observation period (Note: InitialPopulation does not change its value). Let the user input the InitialPopulation, which is the number of bacteria present at the beginning of the trial and the GrowthRate (as a percentage). Then compute the number of bacteria in the culture after each day for the first 10 days (Time will have values 1 through 10). Do this in a loop so the user can see the results in a table (properly format your table with column headers and a title). The output table should have headings for Day and Number of Bacteria Present (on that day). 

NOTE: The number e is a mathematical constant that is the base of the natural logarithm. It is approximately equal to 2.71828 and is available in the C# Math library.

```
int initialPopulation;
double growthRate;

double finalPopulation;

Console.Write("Please enter the initial bacterial population: ");
// initialPopulation = int.Parse(Console.ReadLine());
initialPopulation = 1245;

Console.Write("Please enter the growth rate: \n00.0%\r");
// growthRate = double.Parse(Console.ReadLine()) / 100;
growthRate = 26.7 / 100;

Console.WriteLine($"Initial Bacterial Population = {initialPopulation:##,#}\n" +
    $"Growth Rate = {growthRate:P2}\n\n");

Console.WriteLine($" {"DAY",-4}|{"NUMBER OF BACTERIA",19}");
Console.WriteLine("-----|-------------------");
for (int days = 1; days <= 10; days++)
{
    finalPopulation = initialPopulation * Math.Pow(Math.E, growthRate * days);
    Console.WriteLine($" {days,-4}|{finalPopulation,19:N2}");
} // end of for
```

# Question 2
Write a program that will prompt the user for a string (could be anything), and then prompts the user for a single character. The program should then display a count of the number of times the character entered appears in the string.

```
string userInput;
char userCharacter;
int count = 0;

Console.WriteLine("Please enter a string:");
// userInput = Console.ReadLine();
userInput = "The quick brown fox jumps over the lazy dog";

Console.WriteLine("Please enter a character:");
//userCharacter = char.Parse(Console.ReadLine());
userCharacter = 'o';

foreach (char item in userInput.ToCharArray())
{
    if (item == userCharacter)
    {
        count++;
    } // end of if
} // end of foreach

if (count == 1)
{
    Console.WriteLine($"The character \'{userCharacter}\' " +
        $"appears {count} time in the string.");
} // end of if
else
{
    Console.WriteLine($"The character \'{userCharacter}\' " +
        $"appears {count} times in the string.");
} // end of else
```

# Question 3
Write a program for an Internet service provider so they can 
calculate how much to bill their customers. The provider 
offers customers 3 packages: 

PACKAGE COST CALCULATION 
* A | $9.95 per month for 10 hours; additional hours are billed at $2.00 per hour 
* B | $13.95 per month for 20 hours; additional hours are billed at $1.00 per hour 
* C | $19.95 per month of unlimited hours 

The program should prompt for the letter of the service 
package (A, B, or C) and the number of hours they have used. 
The program should then display the total amount billed. 
The program should allow for multiple bills to be processed 
(i.e. use a loop).

```
char package, userSelection;
int numberOfHours;
double monthlyBill;

do
{
    Console.WriteLine("Which package would you like: \n" +
    "A: $9.95 per month for up to 10 hours; additional hours are billed at $2.00 per hour\n" +
    "B: $13.95 per month for up to 20 hours; additional hours are billed at $1.00 per hour\n" +
    "C: $19.95 per month for unlimited hours\n");

    // package = char.Parse(Console.ReadLine());
    package = 'B';

    if (package == 'a' || package == 'A' || package == 'b' || package == 'B')
    {
        Console.WriteLine("How many full hours have you used on your internet service?\n");
        //numberOfHours = int.Parse(Console.ReadLine());
        numberOfHours = 25;

        switch (package)
        {
            case 'a':
            case 'A':
                if (numberOfHours > 10)
                {
                    monthlyBill = 9.95 + (numberOfHours - 10) * 2;
                } // end of if
                else
                {
                    monthlyBill = 9.95;
                } // end of else
                Console.WriteLine($"Thank you for choosing package {package}\n" +
                                  $"Your bill is {monthlyBill:C}/month");
                break;
            default:
                if (numberOfHours > 20)
                {
                    monthlyBill = 13.95 + (numberOfHours - 20);
                } // end of if
                else
                {
                    monthlyBill = 13.95;
                } // end of else
                Console.WriteLine($"Thank you for choosing package {package}\n" +
                                  $"Your bill is {monthlyBill:C}/month");
                break;
        } // end of switch
    } // end of if
    else
    {
        monthlyBill = 19.95;
        Console.WriteLine($"Thank you for choosing package {package}\n" +
                          $"Your bill is {monthlyBill:C}/month");
    } // end of else

    Console.WriteLine("\n\nTo make another bill enter \'Y\'\n" +
        "To quit enter \'N\'");
    
    // userSelection = char.Parse(Console.ReadLine());
    userSelection = 'n';

} // end of do
while (userSelection == 'y' || userSelection == 'Y');

Console.WriteLine(@"
***************************************************
**********************THANK YOU FOR YOUR BUSINESS**
***************************************************
 ______     ______     ______     ______     ______     ______               
/\  == \   /\  ___\   /\  == \   /\  ___\   /\  ___\   /\  == \              
\ \  __<   \ \  __\   \ \  __<   \ \ \__ \  \ \  __\   \ \  __<              
 \ \_____\  \ \_____\  \ \_\ \_\  \ \_____\  \ \_____\  \ \_\ \_\            
  \/_____/   \/_____/   \/_/ /_/   \/_____/   \/_____/   \/_/ /_/            
                                                                             
 ______   ______     __         ______     ______     ______     __    __    
/\__  _\ /\  ___\   /\ \       /\  ___\   /\  ___\   /\  __ \   /\ ""-./  \   
\/_/\ \/ \ \  __\   \ \ \____  \ \  __\   \ \ \____  \ \ \/\ \  \ \ \-./\ \  
   \ \_\  \ \_____\  \ \_____\  \ \_____\  \ \_____\  \ \_____\  \ \_\ \ \_\ 
    \/_/   \/_____/   \/_____/   \/_____/   \/_____/   \/_____/   \/_/  \/_/");
```
