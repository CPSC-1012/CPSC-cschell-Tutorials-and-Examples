# Question 1

```
using System;

namespace MethodExercisesFour
{
    class Program
    {
        static void Main()
        {
            int pennies = GetPositiveInt("pennies");
            int nickles = GetPositiveInt("nickles");
            int dimes = GetPositiveInt("dimes");
            int quarters = GetPositiveInt("quarters");
            int loonies = GetPositiveInt("loonies");
            int twoonies = GetPositiveInt("twoonies");
            double total = GetTotal(pennies, nickles, dimes, 
                                    quarters, loonies, twoonies);
            
            Console.WriteLine($"Total is {total:C}");
        } // end of method
        public static double GetTotal(int pennies, int nickels, int dimes, 
                                      int quarters, int loonies, int twoonies)
        {
            return pennies * 0.01 + nickels * 0.05 + dimes * 0.1 + 
                   quarters * 0.25 + loonies * 1 + twoonies * 2;
        } // end of method
        public static int GetPositiveInt(string prompt)
        {
            int num;
            Console.Write($"Enter number of {prompt}: ");

            while (!int.TryParse(Console.ReadLine(), out num) || num < 0)
            {
                Console.Write($"Enter a positive number of {prompt}: ");
            } // end of while
            
            return num;

        } // end of method
    } // end of class
} // end of namespace
```

# Question 2

```
using System;

namespace MethodExercisesFive
{
    class Program
    {
        static void Main()
        {
            int seconds = GetPositiveInt("seconds");
            int minutes = GetPositiveInt("minutes");
            int hours = GetPositiveInt("hours");
            int totalSeconds;

            totalSeconds = CalculateSeconds(seconds, minutes, hours);

            Console.WriteLine($"The total is {totalSeconds:N0}s");

        } // end of method
        public static int CalculateSeconds(int seconds, int minutes, int hours)
        {
            int tempseconds = hours * 60 * 60;
            tempseconds += minutes * 60;
            return tempseconds + seconds;
        } // end of method
        public static int GetPositiveInt(string prompt)
        {
            int number;
            Console.Write($"Enter number of {prompt}: ");

            while(!int.TryParse(Console.ReadLine(), out number) || number < 0)
            {
                Console.Write($"Enter a positive number of {prompt}: ");
            } // end of while

            return number;
        } // end of method
    } // end of class
} // end of namespace
```

# Question 3

```
using System;

namespace MethodExercisesSix
{
    class Program
    {
        static void Main()
        {
            double hours = GetPositivedouble("number of hours");
            double rate = GetPositivedouble("pay rate");

            double totalPay = CalculatePay(hours, rate);

            Console.WriteLine($"The total pay is {totalPay:C}");
        } // end of method
        public static double CalculatePay(double hours, double rate)
        {
            double tempPay, tempHours, total;
            double dbltime = 0, halftime = 0;
            if (hours > 50)
            {
                tempHours = hours - 50;
                dbltime = tempHours * (rate * 2);
                hours -= tempHours;
            } // end of if
            if (hours > 40)
            {
                tempHours = hours - 40;
                halftime = tempHours * (rate * 1.5);
                hours -= tempHours;
            } // end of if
            tempPay = hours * rate;
            total = tempPay + halftime + dbltime;
            return total;
        } // end of method
        public static double GetPositivedouble(string prompt)
        {
            double number;
            Console.Write($"Enter the {prompt}: ");

            while (!double.TryParse(Console.ReadLine(), out number) || number < 0)
            {
                Console.Write($"Enter a positive {prompt}: ");
            } // end of while
            return number;
        } // end of method
    } // end of class
} // end of namespace
```

# Question 4

```
using System;

namespace MethodExercisesSeven
{
    class Program
    {
        static void Main()
        {
            int length = GetPositiveInt("length of one side in cm");
            int width = GetPositiveInt("width of one side in cm");
            int height = GetPositiveInt("height in cm");

            double area = CalculateArea(length, width);
            double volume = CalculateVolume(length, width, height);

            Console.WriteLine($"The rectangle has a volume of {volume:N0}cm^3");
            Console.WriteLine($"The rectangle face has an area of {area:N0}cm^2");
        } // end of method
        public static double CalculateArea(int length, int width)
        {
            return length * width;
        } // end of method
        public static double CalculateVolume(int length, int width, int height)
        {
            return length * width * height;
        } // end of method
        public static int GetPositiveInt(string prompt)
        {
            int num;
            Console.Write($"Enter {prompt}: ");

            while (!int.TryParse(Console.ReadLine(), out num) || num < 0)
            {
                Console.Write($"Enter a positive {prompt}: ");
            } // end of while
            return num;
        } // end of method
    } // end of class
} // end of namespace
```

# Question 5

```
using System;

namespace MethodExercisesEight
{
    class Program
    {
        static void Main()
        {
            char menu;
            double total = 0;
            do
            {
                DisplayMenu();
                while (!char.TryParse(Console.ReadLine(), out menu) && 
                      (char.ToUpper(menu) != 'A' || 
                       char.ToUpper(menu) != 'R' || 
                       char.ToUpper(menu) != 'F' || 
                       char.ToUpper(menu) != 'X'))
                {
                    Console.WriteLine("Invalid menu option");
                    DisplayMenu();
                } // end of while
                menu = char.ToUpper(menu);
                switch (menu)
                {
                    case 'A':
                        total += 25;
                        break;
                    case 'R':
                        total += RegularFare();
                        break;
                    case 'F':
                        total += FlatRate();
                        break;
                    default:
                        break;
                } // end of switch
                Console.WriteLine($"Total is {total:C2}");
            } while (menu != 'X');

            Console.WriteLine($"Final total is {total:C2}");
        } // end of method
        public static double FlatRate()
        {
            Console.Write("Enter flat rate: ");
            return GetSafeDouble();
        } // end of method
        public static void DisplayMenu()
        {
            Console.WriteLine("A: Airport Trip $25.00");
            Console.WriteLine("R: Regular Fare ");
            Console.WriteLine("F: Flat Rate");
            Console.WriteLine("X: Exit");
        } // end of while
        public static double RegularFare()
        {
            double total;
            double distance, time;
            Console.Write("Enter distance traveled: ");
            distance = GetSafeDouble();
            Console.Write("Enter length of trip in minutes: ");
            time = GetSafeDouble();

            total = distance * 1.1 + time * 0.2;
            return total;
        } // end of method
        public static double GetSafeDouble()
        {
            double num;
            while (!double.TryParse(Console.ReadLine(), out num) || num < 0)
            {
                Console.WriteLine("Error try again");
            } // end of while
            return num;
        } // end of method
    } // end of class
} // end of namespace
```

# Question 6 -- By Reference

```
using System;

namespace MethodExercisesNineRef
{
    class Program
    {
        static void Main()
        {
            int byValue;
            double byRef1 = 0, byRef2 = 0;
            Console.Write("Enter a number: ");
            byValue = int.Parse(Console.ReadLine());

            GetNumberAboveAndBelow(byValue, ref byRef1, ref byRef2);

            Console.WriteLine("The number before the square root of " + byValue + " is " + byRef1 + " and after is " + byRef2);
            Console.ReadKey();
        } // end of method
        public static void GetNumberAboveAndBelow(int value, ref double byRef1, ref double byRef2)
        {
            double squareRoot = Math.Sqrt(value);
            byRef1 = Math.Floor(squareRoot);
            byRef2 = Math.Ceiling(squareRoot);
        } // end of method
    } // end of class
} // end of namespace
```

# Question 7 -- By Reference

```
using System;

namespace MethodExercisesTenRef
{
    class Program
    {
        static void Main()
        {
            int number1;
            int number2;
            int number3;
            int high = 0;
            int middle = 0;
            int low = 0;

            number1 = GetPositiveInt("Enter the first number", "That is not a valid number. Try again: ");
            number2 = GetPositiveInt("Enter the second number", "That is not a valid number. Try again: ");
            number3 = GetPositiveInt("Enter the third number", "That is not a valid number. Try again: ");

            FindHighMediumLow(ref high, ref low, ref middle, number1, number2, number3);

            Console.Write("High is {0}, Middle is {1}, Low is {2}", high, middle, low);
            Console.ReadKey();

        } // end of method
        public static int GetPositiveInt(string prompt, string error)
        {
            int num;
            Console.Write(prompt + ": ");
            while (!int.TryParse(Console.ReadLine(), out num) || num < 0)
            {
                Console.Write(error);
            } // end of while
            return num;
        } // end of method
        public static void FindHighMediumLow(ref int high, ref int low, ref int middle, int num1, int num2, int num3)
        {
            if (num1 > num2)
            {
                high = num1;
                middle = num2;
                if (num3 > high)
                {
                    high = num3;
                    middle = num1;
                    low = num2;
                } // end of if
                else
                {
                    if (num3 > middle)
                    {
                        middle = num3;
                        low = num2;
                    } // end of if
                    else
                    {
                        low = num3;
                    } // end of else
                } // end of else
            } // end of if
            else
            {
                high = num2;
                middle = num1;
                if (num3 > high)
                {
                    high = num3;
                    middle = num2;
                    low = num1;
                } // end of if
                else
                {
                    if (num3 > middle)
                    {
                        middle = num3;
                        low = num1;
                    } // end of if
                    else
                    {
                        low = num3;
                    } // end of else
                } // end of else
            } // end of else
        } // end of method
    } // end of class
} // end of namespace
```
