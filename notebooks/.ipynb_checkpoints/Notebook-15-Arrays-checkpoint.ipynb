1. Create a program that will allow the user to enter up to 25 numbers into an array. Calculate and display the mean average of the numbers.

````
using System;

namespace ConsoleApp71
{
    class Program
    {
        static void Main()
        {
            const int MAX_LENGTH = 25;
            int[] numbers = new int[MAX_LENGTH];
            int count;
            double average;

            count = PopulateArray(MAX_LENGTH, numbers);
            average = CalculateAverage(numbers, count);

            Console.WriteLine($"The average for the array inputs is {average:F2}");
        }
        public static int PopulateArray(int max, int[] fillArray)
        {
            int input = 0;
            int count = 0;
            Console.WriteLine("Enter numbers to calculate average. \n" +
                "Positive numbers only. \n" +
                "Enter -1 to quit.\n");
            while (count < max && input != -1)
            {
                Console.Write($"Enter number {count + 1}: ");
                while (!int.TryParse(Console.ReadLine(), out input) || input < -2)
                {
                    Console.Write("Invalid number. Try again: ");
                }
                if (input != -1)
                {
                    fillArray[count] = input;
                    count++;
                }
            }
            return count;
        }
        static double CalculateAverage(int[] inputArray, int count)
        {
            int sum = 0;
            foreach (int item in inputArray)
            {
                sum += item;
            }
            return sum / count;
        }
    }
}
````

3. Modify the previous program (make a copy and save as a new project) so thatthe program also displays the largest value, the smallest value, and the mode (if you donâ€™t know what the mode is, look it up) of the numbers. Use a modular approach in your solution. Your program should have at a minimum the following methods:

`Function LargestNumber(Integer[ ] Nums, Integer Count)` As Integer
The method will return the index of the largest number in the array.

`Function SmallestNumber(Integer[ ] Nums, Integer Count)` As Integer
The method will return the index of the smallest number in the array.

`Function Mode(Integer[ ] Nums, Integer Count)` As Integer
The method will return the mode of the numbers (if there is more than one, only one needs to be shown).

````
using System;
using System.Linq;

namespace ConsoleApp71
{
    class Program
    {
        static void Main()
        {
            const int MAX_LENGTH = 25;
            int[] numbers = new int[MAX_LENGTH];
            int count;
            double average;
            int largestNumber;
            int smallestNumber;
            int mode;

            count = GetUserNumbers(numbers);
            average = MeanAverageOfNumbers(numbers, count);
            largestNumber = LargestNumber(numbers, count);
            smallestNumber = SmallestNumber(numbers, count);
            mode = Mode(numbers, count);

            Console.WriteLine($"The average for the array inputs is {average:F2}");
            Console.WriteLine($"The largest number is {numbers[largestNumber]}");
            Console.WriteLine($"The smallest number is {numbers[smallestNumber]}");
            Console.WriteLine($"The mode is {mode}");
        }
        
        public static int GetUserNumbers(int[] numbers)
        {
            int input = 0;
            int count = 0;
            Console.WriteLine("Enter numbers to calculate average. \n" +
                "Positive numbers only. \n" +
                "Enter -1 to quit.\n");
            while (count < numbers.Length && input != -1)
            {
                Console.Write($"Enter number {count + 1}: ");
                while (!int.TryParse(Console.ReadLine(), out input) || input < -2)
                {
                    Console.Write("Invalid number. Try again: ");
                }
                if (input != -1)
                {
                    numbers[count] = input;
                    count++;
                }
            }
            return count;
        }
        
        static double MeanAverageOfNumbers(int[] numbers, int count)
        {
            int sum = 0;
            foreach (int item in numbers)
            {
                sum += item;
            }
            return sum / count;
        }

        static int LargestNumber(int[] numbers, int count)
        {
            int maxValue;
            int valueIndex = 0;
            maxValue = numbers[..(count)].Max();

            for (int i = 0; i < count; i++)
            {
                if (numbers[i] == maxValue)
                {
                    valueIndex = i;
                }
            }
            return valueIndex;
        }
        
        static int SmallestNumber(int[] numbers, int count)
        {
            int minValue;
            int valueIndex = 0;
            minValue = numbers[..(count)].Min();

            for (int i = 0; i < count; i++)
            {
                if (numbers[i] == minValue)
                {
                    valueIndex = i;
                }
            }
            return valueIndex;
        }
        
        static int Mode(int[] numbers, int count)
        {
            // https://stackoverflow.com/questions/5069687/find-character-with-most-occurrences-in-string
            int mode = numbers[..(count - 1)].GroupBy(x => x).OrderByDescending(x => x.Count()).First().Key;
            return mode;
        }
    }
}
````

4. Create a program that will allow the user to enter marks for a quiz. The user will need the ability to enter the following:
- The total for the quiz
- Up to, but no more than, 25 quiz marks and corresponding student names

Once the data has been entered, display a menu of options to the user to allow for the following:
- View all marks
- View the highest mark
- View the lowest mark
- Find the mean average of the marks
- Quit the program
Each of the three view options should display the data as a table with appropriate headings. The remaining options are left for you to implement as
you see fit.

Example view printout:
```
    Quiz Total: 25
    Student Name      Mark
    ------------      ----
    Sally Anne         23
    Jon Doe            19
    Mary Jane          21
```

````
/*
    Purpose:    The program tallies marks for a quiz. The user should
                have the ability to perform various actions on the 
                collection. Namely, find the highest, lowest, and average
                mark for the quiz. There is also the ability to save
                and read the data to or from a file respectively.

    Inputs:     Up to 25 marks for 25 students (can be imported via file)
                The quiz total

    Outputs:    Lowest mark achieved
                Highest mark achieved
                Average of all marks
                Data output to file

    Algorithm:  Prompt the user for the quiz total
                Prompt the user for data entry
                do
                    Display the menu
                    Collect and validate the user selection
                        View all marks
                        View the highest mark
                        View the lowest mark
                        Mean of marks
                        Remove mark
                        Save to file
                        Read from file
                        Quit program
                while the user selection is not 0

    Test Plan:
        Test Case       Test Data           Expected Results
        ---------       ---------           ----------------
        Class Results   Quiz total: 5       average     50%
                        Name    Grade       highest     a   5
                        ====    =====       lowest      f   0
                        a       5           file accessible at path
                        b       4           updates at path seen on inport
                        c       3           all values validated and accurate
                        d       2
                        e       1
                        f       0

                        Remove 3            Name    Grade
                        ========            ====    =====
                                            a       5
                                            b       4
                                            d       2
                                            e       1
                                            f       0

    Written By:     Cody Schellenberger

    Written For:    CPSC1012

    Section Number: OA0X

    Last Modified:  18/11/2021
*/
using System;
using System.Linq;
using System.IO;

namespace ConsoleApp72
{
    class Program
    {
        static void Main()
        {
            // set the physical array size and declare the parallel arrays
            const int PHYSICAL_ARRAY_SIZE = 25;
            string[] studentNames = new string[PHYSICAL_ARRAY_SIZE];
            int[] studentGrades = new int[PHYSICAL_ARRAY_SIZE];

            // set the menu choices and quiz parameters
            char[] menuChoices = { '1', '2', '3', '4', '5', '6', '7', '0' };
            char userChoice;
            int logicalArraySize;
            int quizTotal;

            Console.Write("What is the total number of potential marks for the quiz?: ");
            quizTotal = GetSafePositiveInteger();

            // capture the user inputs and set the logical size for the arrays
            logicalArraySize = GetStudentGrades(studentNames, studentGrades, quizTotal);

            // pocess the menu choices using a loop
            do
            {
                userChoice = GetMenuChoice(menuChoices, quizTotal);
                ProcessMenuChoice(userChoice, studentGrades, studentNames, ref logicalArraySize, quizTotal, PHYSICAL_ARRAY_SIZE);
            } while (userChoice != menuChoices[^1]);
        }
        public static int GetStudentGrades(string[] names, int[] grades, int quizTotal)
        {
            int input = 0;
            string name;
            int count = 0;
            Console.WriteLine("Enter quiz grades \n" +
                "    - Positive numbers only. \n" +
                "    - Enter a quiz grade of -1 to quit.\n");
            while (count < names.Length && input != -1)
            {
                string message = "Enter the student name: ";
                name = PromptForStringNotEmpty(message);
                
                Console.Write($"Enter the grade for {name}: ");
                while (!int.TryParse(Console.ReadLine(), out input) || input < -2 || input > quizTotal)
                {
                    Console.Write($"Invalid number. Please enter a number between 0 and {quizTotal} (inclusive): ");
                }
                if (input != -1)
                {
                    grades[count] = input;
                    names[count] = name;
                    count++;
                }
            }
            return count;
        }
        static void DisplayMainMenu(int quizTotal)
        {
            Console.WriteLine(" ______________________________");
            Console.WriteLine("|  __________________________  |");
            Console.WriteLine("| |                          | |");
            Console.WriteLine($"| | Quiz Total: {quizTotal,-12} | |");
            Console.WriteLine("| | ------------------------ | |");
            Console.WriteLine("| | 1. View all marks        | |");
            Console.WriteLine("| | 2. View the highest mark | |");
            Console.WriteLine("| | 3. View the lowest mark  | |");
            Console.WriteLine("| | 4. Mean of marks         | |");
            Console.WriteLine("| | 5. Remove mark           | |");
            Console.WriteLine("| | 6. Save to file          | |");
            Console.WriteLine("| | 7. Read from file        | |");
            Console.WriteLine("| | 0. Quit program          | |");
            Console.WriteLine("| |__________________________| |");
            Console.WriteLine("|______________________________|");
            Console.Write("Enter your choice: ");
        }
        static bool IsValidMenuChoice(char[] validMenuChoices, char selectedMenuChoice)
        {
            bool validChoice = false;
            for (int index = 0; index < validMenuChoices.Length; index++)
            {
                if (selectedMenuChoice == validMenuChoices[index])
                {
                    validChoice = true;
                    index = validMenuChoices.Length;
                }
            }
            return validChoice;
        }
        static char GetMenuChoice(char[] menuChoices, int quizTotal)
        {
            char userChoice;
            bool validChoice = false;

            do
            {
                DisplayMainMenu(quizTotal);
                userChoice = Console.ReadKey().KeyChar;
                Console.WriteLine();
                if (!IsValidMenuChoice(menuChoices, userChoice))
                {
                    Console.ForegroundColor = ConsoleColor.Red;
                    Console.WriteLine("Invalid menu choice. Try again.");
                    Console.WriteLine("Press any key to continue...");
                    Console.ForegroundColor = ConsoleColor.Gray;
                    Console.ReadKey();
                    Console.Clear();
                }
                else
                {
                    validChoice = true;
                }

            } while (!validChoice);

            return userChoice;
        }

        static void ViewAllMarks(int quizTotal, string[] studentNames, int[] studentGrades, int count)
        {
            Console.WriteLine($"Quiz Total: {quizTotal}");
            Console.WriteLine($"Student Name          Mark");
            Console.WriteLine($"------------          ----");
            for (int i = 0; i < count; i++)
            {
                Console.WriteLine($"{studentNames[i],-23}{studentGrades[i],-3}");
            }
        }

        static void ViewHighestMark(int quizTotal, string[] studentNames, int[] studentGrades, int count)
        {
            int maxValue;
            int valueIndex = 0;
            maxValue = studentGrades[..count].Max();

            for (int i = 0; i < count; i++)
            {
                if (studentGrades[i] == maxValue)
                {
                    valueIndex = i;
                }
            }
            Console.WriteLine("Highest Mark Lookup");
            Console.WriteLine("===================");
            Console.WriteLine($"Quiz Total: {quizTotal}");
            Console.WriteLine($"Student Name          Mark");
            Console.WriteLine($"------------          ----");
            Console.WriteLine($"{studentNames[valueIndex],-23}{studentGrades[valueIndex],-3}");
        }

        static void ViewLowestMark(int quizTotal, string[] studentNames, int[] studentGrades, int count)
        {
            int minValue;
            int valueIndex = 0;
            minValue = studentGrades[..count].Min();

            for (int i = 0; i < count; i++)
            {
                if (studentGrades[i] == minValue)
                {
                    valueIndex = i;
                }
            }
            Console.WriteLine("Lowest Mark Lookup");
            Console.WriteLine("===================");
            Console.WriteLine($"Quiz Total: {quizTotal}");
            Console.WriteLine($"Student Name          Mark");
            Console.WriteLine($"------------          ----");
            Console.WriteLine($"{studentNames[valueIndex],-23}{studentGrades[valueIndex],-3}");
        }

        static double FindMeanPercent(int[] studentGrades, int count, int quizTotal)
        {
            int sum = 0;
            foreach (int grade in studentGrades)
            {
                sum += grade;
            }

            return ((double)sum / count) / quizTotal;
        }

        static void ProcessMenuChoice(char userChoice, int[] studentGrades, string[] studentNames, ref int logicalSize, int quizTotal, int physicalSize)
        {
            switch (userChoice)
            {
                case '1':
                    {
                        ViewAllMarks(quizTotal, studentNames, studentGrades, logicalSize);
                    }
                    break;
                case '2':
                    {
                        ViewHighestMark(quizTotal, studentNames, studentGrades, logicalSize);
                    }
                    break;
                case '3':
                    {
                        ViewLowestMark(quizTotal, studentNames, studentGrades, logicalSize);
                    }
                    break;
                case '4':
                    {
                        Console.WriteLine($"The average percentage is {FindMeanPercent(studentGrades, logicalSize, quizTotal):P2}");
                    }
                    break;
                case '5':
                    {
                        RemoveItem(studentNames, studentGrades, ref logicalSize);
                    }
                    break;
                case '6':
                    {
                        SaveToFile(studentNames, studentGrades, logicalSize);
                    }
                    break;
                case '7':
                    {
                        logicalSize = LoadFromFile(studentNames, studentGrades, physicalSize);
                    }
                    break;
                case '0':
                    Console.ForegroundColor = ConsoleColor.Yellow;
                    Console.WriteLine("Good-bye.");
                    break;
                default:
                    break;
            }
            if (userChoice != '0')
            {
                Console.WriteLine("Press any key to continue...");
                Console.ForegroundColor = ConsoleColor.Gray;
                Console.ReadKey();
                Console.Clear();
            }
        }

        static bool RemoveItem(string[] studentNames, int[] studentGrades, ref int logicalSize)
        {
            bool removed = false;

            if (logicalSize == 0)
            {
                Console.WriteLine("There are no marks to remove.");
            }
            else
            {
                Console.Clear();
                Console.WriteLine($"{"Entry",6}  {"Student Name",-23} {"Mark",-3}");
                Console.WriteLine($"{"------",6}  {"-----------",-23} {"------",-3}");
                for (int index = 0; index < logicalSize; index++)
                {
                    Console.WriteLine($"{index + 1,6}  {studentNames[index],-23} {studentGrades[index],-3}");
                }
                //WriteLine("Which item do you like to remove?");
                int removeMarkIndex = PromptForIntegerRangeValue("Which entry number would you like to remove? 0 to cancel: ", 0, logicalSize);
                if (removeMarkIndex == 0)
                {
                    Console.WriteLine("No mark has been removed.");
                }
                else
                {
                    removeMarkIndex -= 1;
                    // Shuffle all the items up 
                    for (int index = removeMarkIndex; index < logicalSize - 1; index++)
                    {
                        studentNames[index] = studentNames[index + 1];
                        studentGrades[index] = studentGrades[index + 1];
                    }
                    studentNames[logicalSize - 1] = null;
                    studentGrades[logicalSize - 1] = 0;
                    logicalSize -= 1;
                    Console.WriteLine("Remove item was successful.");
                    removed = true;
                }

            }
            return removed;
        }
        
        public static int GetSafePositiveInteger()
        {
            bool validInput = int.TryParse(Console.ReadLine(), out int userInput);
            while (!validInput || userInput < 0)
            {
                Console.Write("Invalid input. Please enter a valid positive integer: ");
                validInput = int.TryParse(Console.ReadLine(), out userInput);
            } // end of while
            return userInput;
        }

        static int PromptForIntegerRangeValue(string message, int minValue, int maxValue)
        {
            int integerValue = 0;
            bool validInput = false;
            string userInput;

            do
            {
                Console.Write(message);
                userInput = Console.ReadLine();
                try
                {
                    integerValue = int.Parse(userInput);
                    if (integerValue >= minValue && integerValue <= maxValue)
                    {
                        validInput = true;
                    }
                    else
                    {
                        Console.ForegroundColor = ConsoleColor.Red;
                        Console.WriteLine($"Invalid input value. Please enter a integer value between {minValue} and {maxValue}.");
                        Console.ForegroundColor = ConsoleColor.Gray;
                    }
                }
                catch (FormatException)
                {
                    Console.ForegroundColor = ConsoleColor.Red;
                    Console.WriteLine("Invalid input value. Please enter a numeric value");
                    Console.ForegroundColor = ConsoleColor.Gray;
                }
            } while (!validInput);

            return integerValue;
        }

        static int LoadFromFile(string[] studentNames, int[] studentGrades, int arrayPhysicalSize)
        {
            int arrayLogicalSize = 0;

            const char delimiter = ',';
            // Prompt and read filename
            Console.WriteLine("Enter the file path to load items from: ");
            string dataImportFilePath = Console.ReadLine();
            try
            {
                using (StreamReader reader = new StreamReader(dataImportFilePath))
                {
                    string line;
                    int index = 0;
                    while (!reader.EndOfStream && index < arrayPhysicalSize)
                    {
                        line = reader.ReadLine();
                        string[] lineArray = line.Split(delimiter);
                        studentNames[index] = lineArray[0];
                        studentGrades[index] = int.Parse(lineArray[1]);
                        index++;
                    }
                    arrayLogicalSize = index;
                    Console.WriteLine($"Read from {dataImportFilePath} was successful.");
                } 
            }
            catch (Exception ex)
            {
                Console.WriteLine(ex.Message);
            }

            return arrayLogicalSize;
        }

        static void SaveToFile(string[] studentNames, int[] studentMarks, int logicalArraySize)
        {
            const char delimiter = ',';
            // Prompt and read filename
            Console.WriteLine("Enter the file path to save items to: ");
            string dataExportFilePath = Console.ReadLine();
            try
            {
                using (StreamWriter writer = new(dataExportFilePath))
                {
                    for (int index = 0; index < logicalArraySize; index++)
                    {
                        writer.WriteLine($"{studentNames[index]}{delimiter}{studentMarks[index]}");
                    }
                    Console.WriteLine($"Write to file {dataExportFilePath} was successful.");
                }
            }
            catch (Exception ex)
            {
                Console.WriteLine(ex.Message);
            }
        }

        static string PromptForStringNotEmpty(string message)
        {
            string stringValue;
            bool validInput = false;

            do
            {
                Console.Write(message);
                stringValue = Console.ReadLine();
                if (!string.IsNullOrEmpty(stringValue))
                {
                    validInput = true;
                }
                else
                {
                    Console.ForegroundColor = ConsoleColor.Red;
                    Console.WriteLine("Invalid input value. Please enter a non-empty string.");
                    Console.ForegroundColor = ConsoleColor.Gray;
                }
            } while (!validInput);

            return stringValue;
        }
    }
}
````
