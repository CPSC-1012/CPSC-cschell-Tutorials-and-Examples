# Question 1
* Write a program for an Internet service provider so they can calculate how much to bill their customers. The provider offers customers 3 packages:
| Package 	| Cost Calculation                                                                   	|
|---------	|------------------------------------------------------------------------------------	|
| A       	| \$9.95 per month for up to 10 hours; additional hours are billed at \$2.00 per hour  	|
| B       	| \$13.95 per month for up to 20 hours; additional hours are billed at \$1.00 per hour 	|
| C       	| \$19.95 per month for unlimited hours                                               	|
* The program should prompt for the letter of the service package (A, B, or C) and the number of hours they have used if either option A or B has been selected.

```
char package;
int numberOfHours;
double monthlyBill;

Console.WriteLine("Which package would you like: \n" +
    "A: $9.95 per month for up to 10 hours; additional hours are billed at $2.00 per hour\n" +
    "B: $13.95 per month for up to 20 hours; additional hours are billed at $1.00 per hour\n" +
    "C: $19.95 per month for unlimited hours\n");

// package = char.Parse(Console.ReadLine());
package = 'W';

if (package == 'a' || package == 'A' || package == 'b' || package == 'B' || package == 'c' || package == 'C'  )
{
    if (package == 'a' || package == 'A' || package == 'b' || package == 'B')
    {
        Console.WriteLine("How many full hours have you used on your internet service?\n");
        // numberOfHours = int.Parse(Console.ReadLine());
        numberOfHours = 25;
        switch (package)
        {
            case 'a':
            case 'A':
                if (numberOfHours > 10)
                {
                    monthlyBill = 9.95 + (numberOfHours - 10) * 2;
                } // end of if
                else
                {
                    monthlyBill = 9.95;
                } // end of else
                Console.WriteLine($"Thank you for choosing package {package}\n" +
                                  $"Your bill is {monthlyBill:C}/month");
                break;
            default:
                if (numberOfHours > 20)
                {
                    monthlyBill = 13.95 + (numberOfHours - 20);
                } // end of if
                else
                {
                    monthlyBill = 13.95;
                } // end of else
                Console.WriteLine($"Thank you for choosing package {package}\n" +
                                  $"Your bill is {monthlyBill:C}/month");
                break;
        } // end of switch
    } // end of if
    else
    {
        monthlyBill = 19.95;
        Console.WriteLine($"Thank you for choosing package {package}\n" +
                          $"Your bill is {monthlyBill:C}/month");
    } // end of else
} // end of if
else
{
    Console.WriteLine($"That isn't a package. Please try again.");
}
    
```

# Question 2
* Write a program that can allow the user to perform some simple unit conversions. The program should display a menu of available options to the user, prompt for the desired option, and then prompt for the myValue. The conversion should be performed and the new myValue should be displayed. Use the following table for conversions:

| Conversion                    	| Formula                     	|
|-------------------------------	|-----------------------------	|
| Pounds to Kilograms           	| kg = lb / 2.2046            	|
| Fluid Ounces (U.S.) to Liters 	| 1 liter = 33.8140226 ounces 	|
| Inches to Centimeters         	| cm = in / 0.39370           	|
| Fahrenheit to Celsius         	| C = (F - 32) / 1.800        	|

* As an option, also allow for conversions in the opposite direction (e.g. kilograms to pounds)

```
int choice;
double myValue;
double answer;

Console.WriteLine("Hello to the conversion application!\n" +
    "Please select the conversion you would like to compute:\n\n" +
    "(1) Pounds to Kilograms\n" +
    "(2) Fluid Ounces (U.S.) to Liters\n" +
    "(3) Inches to Centimeters\n" +
    "(4) Fahrenheit to Celsius\n\n" +
    "Please input the number:");
/*
choice = int.Parse(Console.ReadLine());
*/
choice = 3;

Console.WriteLine("What value do you want to convert?");
/*
myValue = double.Parse(Console.ReadLine());
*/
myValue = 65.2;

switch (choice)
{
    case 1:
        answer = myValue / 2.2046;
        Console.WriteLine($"{myValue} pound(s) is equal to {answer:N1} kilogram(s)");
        break;
    case 2:
        answer = myValue / 33.8140226;
        Console.WriteLine($"{myValue} ounce(s) is equal to {answer:N4} liter(s)");
        break;
    case 3:
        answer = myValue / 0.39370;
        Console.WriteLine($"{myValue} inch(es) is equal to {answer:##,#0.0} centimeter(s)");
        break;
    case 4:
        answer = (myValue - 32) / 1.800;
        Console.WriteLine($"{myValue} fahrenheit is equal to {answer:N1} celsius");
        break;
    default:
        Console.WriteLine($"{choice} is not a valid choice, please try again.");
        break;
} // end of switch
```
